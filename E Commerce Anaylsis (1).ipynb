# MATPLOTLIB OPERATION TASK:

# 1.Plot a bar chart showing total sales by category.

```python
Data_Of_Category=data.groupby(data['Category']).agg(
    {
        'Sales':'sum',
       
    }
)
categories=Data_Of_Category.index
total_sales=Data_Of_Category['Sales']
```

```python
plt.figure(figsize=(8,5))
plt.bar(categories, total_sales, color=['yellow','red','skyblue'], edgecolor='black', )
plt.title(' Total Sales by Category', fontsize=14, fontweight='bold',)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=20, fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 3.“Plot a pie chart showing percentage of total sales by category.”

```python
plt.figure(figsize=(5,5))
plt.pie(
    total_sales,
    labels=categories,
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'],
    autopct='%1.1f%%'           )
plt.title('Total Sales Distribution by Category')
plt.axis('equal') 
plt.show()
```

# 3.Plot a line chart showing monthly profit trend

```python
data.head()
```

```python
import calendar
monthly_profit=data.groupby('month')['Profit'].sum()
months = monthly_profit.index
profits = monthly_profit.values
month_names = [calendar.month_abbr[m] for m in months]
plt.figure(figsize=(8,5))
plt.plot(month_names,profits,color='lime',marker='o')
plt.xticks(range(1,13))
plt.ticklabel_format(style='plain', axis='y')
plt.grid(True,alpha=0.2)
plt.xlabel('Month')
plt.ylabel("Profit")
plt.title('monthly profit trend')
plt.show()
```

# 4.Plot a scatter plot between Sales_Amount and Profit to observe correlation.

```python
sales=data['Sales']
profit=data['Profit']
plt.scatter(sales,profit,color='black',alpha=0.5)
plt.xlabel("Sales Amount")
plt.ylabel("Profit")
plt.grid(True,alpha=0.7)
plt.title('Scatter Chart between Sales_Amount and Profit')
plt.show()
```
